```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
N = 101
x = np.linspace(-2, 2, N)
y = np.linspace(0, 0.5, N)
X, Y = np.meshgrid(x, y)

def func1(x,y):
    z = x - (1-y)/(x+y)**2 + (y)/(x+y-1)**2
    return z

def func2(x,y):
    z = x - (1-y)/(x+y)**2 - (y)/(x+y-1)**2
    return z

def func3(x,y):
    z = x + (1-y)/(x+y)**2 + (y)/(x+y-1)**2
    return z

Z1 = func1(X,Y)
Z2 = func2(X,Y)
Z3 = func3(X,Y)

plt.figure(figsize=(14,8))
plt.contour(X, Y, Z1, [0], colors = 'purple')
plt.contour(X, Y, Z2, [0], colors = 'blue')
plt.contour(X, Y, Z3, [0], colors = 'red')
    
plt.xlabel(r'$\xi$', fontsize = 25)
plt.ylabel(r'$\mu$', fontsize = 25)
colors = ['purple', 'blue', 'red']
lines = [Line2D([0], [0], color=c, linewidth=3, linestyle='-') for c in colors]
labels = [r'$L_1$', r'$L_2$', r'$L_3$']
plt.legend(lines, labels, fontsize = 20, loc = 'upper left')

plt.grid()
plt.xlim(-2,2)
plt.ylim(0,0.5)
plt.xticks(np.linspace(-2, 2, 21))
plt.savefig("Lagrange_points.jpg")
plt.show()
```
